```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Datos proporcionados
metros = np.array([1, 2, 3, 4, 5, 6, 7, 8])
ni√±os = np.array([2, 6, 10, 5, 10, 3, 2, 2])

# a) Tabla de Frecuencia
df = pd.DataFrame({'Metros': metros, 'Ni√±os': ni√±os})

# Frecuencias absolutas
frecuencia_absoluta = df['Ni√±os']

# Frecuencias relativas
frecuencia_relativa = df['Ni√±os'] / df['Ni√±os'].sum()

# Frecuencia acumulada
frecuencia_acumulada = frecuencia_absoluta.cumsum()

# A√±adir las frecuencias al DataFrame
df['Frecuencia Absoluta'] = frecuencia_absoluta
df['Frecuencia Relativa'] = frecuencia_relativa
df['Frecuencia Acumulada'] = frecuencia_acumulada

print("Tabla de Frecuencia:")
print(df)

# Diagramas de barras
plt.figure(figsize=(14, 7))

# Diagrama de Barras para las frecuencias absolutas
plt.subplot(1, 3, 1)
plt.bar(metros, frecuencia_absoluta, color='skyblue')
plt.xlabel('Metros')
plt.ylabel('Frecuencia Absoluta')
plt.title('Frecuencia Absoluta')

# Diagrama de Barras para las frecuencias relativas
plt.subplot(1, 3, 2)
plt.bar(metros, frecuencia_relativa, color='lightgreen')
plt.xlabel('Metros')
plt.ylabel('Frecuencia Relativa')
plt.title('Frecuencia Relativa')

# Diagrama de Barras para las frecuencias acumuladas
plt.subplot(1, 3, 3)
plt.bar(metros, frecuencia_acumulada, color='lightcoral')
plt.xlabel('Metros')
plt.ylabel('Frecuencia Acumulada')
plt.title('Frecuencia Acumulada')

plt.tight_layout()
plt.show()

# b) Mediana, media aritm√©tica, moda y cuartiles
# Expandir los datos para calcular estad√≠sticas
datos_expandido = np.repeat(metros, ni√±os)

media = np.mean(datos_expandido)
mediana = np.median(datos_expandido)
moda = stats.mode(datos_expandido)[0][0]
cuartiles = np.percentile(datos_expandido, [25, 50, 75])

print(f"Media Aritm√©tica: {media}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")
print(f"Cuartiles: {cuartiles}")

# c) Varianza y desviaci√≥n est√°ndar
varianza = np.var(datos_expandido, ddof=1)
desviacion_estandar = np.std(datos_expandido, ddof=1)

print(f"Varianza: {varianza}")
print(f"Desviaci√≥n Est√°ndar: {desviacion_estandar}")

# d) ¬øEntre qu√© dos valores se encuentra, como m√≠nimo, el 75% de las observaciones?
# Usando el teorema de Chebyshev: al menos el 75% de los datos est√°n dentro de k = 2 desviaciones est√°ndar de la media
k = 2
limite_inferior = media - k * desviacion_estandar
limite_superior = media + k * desviacion_estandar

print(f"El 75% de las observaciones se encuentra, como m√≠nimo, entre {limite_inferior} y {limite_superior}")
```
